# Regresion Lineal Simple

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #Buenos graficos 3D
```

## Generamos datos aleatorios para entrenar

```python
observations = 1000
xs = np.random.uniform(low=-10,high=10,size=(observations,1))
zs = np.random.uniform(-10,10,(observations,1))
inputs = np.column_stack((xs,zs)) #Arregla los vectores columna xs y zs en una matriz de dos columnas

print(inputs.shape)
```

Una matriz de dos columnas y 1000 filas

## Creamos los objetivos

```python
noise = np.random.uniform(-1,1,(observations,1)) #Ruido en los datos
targets = 2*xs - 3*zs + 5 + noise #Funcion objetivo
print(targets.shape)
```

## Grafico de los datos de entrenamiento

```python
targets = targets.reshape(observations,1)
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot(xs,zs,targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Target')
ax.view_init(azim=100)
plt.show()
targets=targets.reshape(observations,1)
```

### Obtiene un punto al azar donde comenzara el entrenamiento en un rango de -0.1 y 0.1

```python
init_range = 0.1
weights = np.random.uniform(-init_range,init_range,size=(2,1))
biases = np.random.uniform(-init_range,init_range,size=1)
print(weights)
print(biases)
```

## Taza de Aprendizaje

```python
learning_rate = 0.02
```

# Entrenamiento del modelo

```python
for i in range(100):
    outputs = np.dot(inputs,weights) + biases #Salidas, ecuacion lineal
    deltas = outputs - targets #Deltas de error
    
    loss = np.sum(deltas**2)/2/observations #perdida media por observacion
    print(loss)
    
    deltas_scaled = deltas/observations
    weights = weights - learning_rate*np.dot(inputs.T,deltas_scaled) #Se usa la transpuesta de las entradas para que se pueda hacer la multiplocacion de mtarices
    biases = biases - learning_rate*np.sum(deltas_scaled)
```

```python
print(weights)
print(biases)
```
