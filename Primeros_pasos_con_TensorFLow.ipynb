```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

## Creamos un conjunto de datos aleatorios y una funcion objetivo. Luego, guardamos los datos en un tipo de archivo amigable con TensorFLow 'npz'

```python
observaciones = 1000
xs = np.random.uniform(low=-10,high=10,size=(observaciones,1))
zs = np.random.uniform(-10,10,(observaciones,1))\

entradas_generadas = np.column_stack((xs,zs))
ruido = np.random.uniform(-1,1,(observaciones,1))
objetivos_generados = 2*xs -3*zs + 5 + ruido

np.savez('TF_intro',entradas=entradas_generadas,objetivos=objetivos_generados)
```

# Se cargan los datos

```python
datos_entrenamiento = np.load('TF_intro.npz')
```

### Se crean dos variables donde se guarden el tamaño de las salidas y las entradas, al tener dos variables las entradas serán 2 y al tener una sola salida o variable dependeiete 'y' será 1. Luego se crea el modelo utilizando la paqueteria de tensorflow y se entrena con los datos de entrenamiento

```python
input_size = 2
output_size = 1

modelo = tf.keras.Sequential([
                              tf.keras.layers.Dense(output_size) #este metodo toma las entradas del modelo y calcula el producto escalar de las entradas y los pesos y suma el sesgo
                              ])

modelo.compile(optimizer='sgd',loss='mean_squared_error')
modelo.fit(datos_entrenamiento['entradas'],datos_entrenamiento['objetivos'],epochs=100,verbose=2)
```

### Una vez entrenado el modelo obtenemos los valores que se le asignaron a las variables despues de las 100 epocas del entrenamiento

## Ahora compararemos los datos originales con los obtenidos con nuestro modelo

```python
modelo.predict_on_batch(datos_entrenamiento['entradas']).round(1)
```

# Grafica de los datos del modelo contra los originales

```python
plt.plot(np.squeeze(modelo.predict_on_batch(datos_entrenamiento['entradas'])),np.squeeze(datos_entrenamiento['objetivos']))
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
